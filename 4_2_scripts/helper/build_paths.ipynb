```python



# Generate paths
paths = build_paths('limited_extent')

# Grouped lists
grouped = {}
for key in paths:
    if '_' in key:
        suffix = key.split('_', 1)[1]
        grouped.setdefault(suffix, []).append(key)

# Write to file
import os
with open(os.path.join('..', '..', '4_1_modules', 'config', 'paths.py'), 'w') as f:
    f.write("import os\nfrom .config import DATA\nfrom dotenv import load_dotenv\nload_dotenv()\nEXTENT = os.getenv('EXTENT')\n\n")
    for var, path in paths.items():
        f.write(f"{var} = {path}\n")
    f.write("\n# Grouped path lists\n")
    for suffix, keys in grouped.items():
        f.write(f"{suffix} = [{', '.join(keys)}]\n")
        
from config.build_paths import build_paths


# Generate paths
paths = build_paths('full_extent', 'FE_')

# Grouped lists
grouped = {}
for key in paths:
    if '_' in key:
        if key.startswith('FE_') and '_' in key[3:]:
            suffix = 'FE_' + key[3:].split('_', 1)[1]
            grouped.setdefault(suffix, []).append(key)


# Write to file
import os
with open(os.path.join('..', '..', '4_1_modules', 'config', 'FE_paths.py'), 'w') as f:
    f.write("import os\nfrom .config import DATA\nfrom dotenv import load_dotenv\nload_dotenv()\nEXTENT = os.getenv('EXTENT')\n\n")
    for var, path in paths.items():
        f.write(f"{var} = {path}\n")
    f.write("\n# Grouped path lists\n")
    for suffix, keys in grouped.items():
        f.write(f"{suffix} = [{', '.join(keys)}]\n")
```
