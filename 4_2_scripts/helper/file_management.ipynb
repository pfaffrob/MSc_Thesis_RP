```python
import os
import shutil

reserves = ['KAU', 'ESK', 'BUS', 'HAM', 'WAI']

for r in use(reserves):
    for src_attr in ['FE_R_MS', 'FE_R_RGB']:
        src_path = os.path.abspath(getattr(r, src_attr))

        # Find the index of '3_0_data' in the path
        try:
            idx = src_path.lower().index('3_0_data')
        except ValueError:
            print(f"'3_0_data' not found in path: {src_path}")
            continue

        # Build the relative path from '3_0_data' onwards
        rel_path = src_path[idx:]
        dst_path = os.path.join('F:\\', rel_path)

        # Copy the entire folder tree
        if os.path.exists(src_path):
            if not os.path.exists(dst_path):
                shutil.copytree(src_path, dst_path)
            else:
                # If destination exists, copy contents individually
                for item in os.listdir(src_path):
                    s = os.path.join(src_path, item)
                    d = os.path.join(dst_path, item)
                    if os.path.isdir(s):
                        shutil.copytree(s, d, dirs_exist_ok=True)
                    else:
                        shutil.copy2(s, d)
        else:
            print(f"Source path does not exist: {src_path}")
```
