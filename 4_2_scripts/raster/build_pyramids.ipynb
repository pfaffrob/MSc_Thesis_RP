```python
from osgeo import gdal
import psutil
import time

# Path to your GeoTIFF file
input_tif = fr"d:\kqn7759\OneDrive - AUT University\privet\privet_detection clean\data\predicted_privet_weights_2000.tif"

# Overview levels to build
overview_levels = [2, 4, 8, 16, 32, 64, 128, 256]

# Get available memory (in MB), use 1/3 of it safely
available_mem_mb = int(psutil.virtual_memory().available / (1024 * 1024) / 3)

# Open the dataset in READ-ONLY mode to force external overviews
ds = gdal.Open(input_tif, gdal.GA_ReadOnly)

# Benchmarking results
results = {}


# Memory-optimized
print(f"Building pyramids (memory-optimized with {available_mem_mb} MB cache)...")
gdal.SetConfigOption("GDAL_NUM_THREADS", "ALL_CPUS")
gdal.SetConfigOption("GDAL_CACHEMAX", str(available_mem_mb))
start = time.time()
ds.BuildOverviews("AVERAGE", overview_levels)
results['memory-optimized'] = time.time() - start

# Clean up
ds = None

# Print results
for config, duration in results.items():
    print(f"Time taken ({config}): {duration:.2f} seconds")
```
